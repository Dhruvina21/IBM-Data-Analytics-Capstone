### Objective: Collect job postings data for various technologies and locations using the Jobs API

```python
import requests
import pandas as pd
import json
import openpyxl
```

```python
# Fetch all data from the API
api_url = "http://127.0.0.1:5000/data/all"

response = requests.get(api_url)
data = response.json()

print("Total job postings fetched:", len(data))
print("\nSample record:")
print(data[0])
```

```python
def get_number_of_jobs_T(technology):
    number_of_jobs = 0
    response = requests.get("http://127.0.0.1:5000/data",
                            params={"Key Skills": technology})
    data = response.json()
    number_of_jobs = len(data)
    return technology, number_of_jobs

# Test with Python
get_number_of_jobs_T("Python")
```

```python
def get_number_of_jobs_L(location):
    number_of_jobs = 0
    response = requests.get("http://127.0.0.1:5000/data",
                            params={"Location": location})
    data = response.json()
    number_of_jobs = len(data)
    return location, number_of_jobs

# Test with Los Angeles
get_number_of_jobs_L("Los Angeles")
```

```python
# List of all technologies
technologies = ["C", "C#", "C++", "Java", "JavaScript", "Python", 
                "Scala", "Oracle", "SQL Server", "MySQL Server", 
                "PostgreSQL", "MongoDB"]

# Collect job counts for each technology
tech_results = []
for tech in technologies:
    tech_results.append(get_number_of_jobs_T(tech))

print("Technology Job Counts:")
for result in tech_results:
    print(f"  {result[0]}: {result[1]}")
```

```python
# List of all locations
locations = ["Los Angeles", "New York", "San Francisco", 
             "Washington DC", "Seattle", "Austin", "Detroit"]

# Collect job counts for each location
location_results = []
for location in locations:
    location_results.append(get_number_of_jobs_L(location))

print("Location Job Counts:")
for result in location_results:
    print(f"  {result[0]}: {result[1]}")
```

```python
# Create Excel workbook
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Job Postings"

# Write headers
ws.append(["Technology/Location", "Number of Jobs", "Category"])

# Write technology data
for tech, count in tech_results:
    ws.append([tech, count, "Technology"])

# Write location data
for location, count in location_results:
    ws.append([location, count, "Location"])

# Save the file
wb.save("job-postings.xlsx")
print("job-postings.xlsx saved successfully!")

# Preview the data as a DataFrame
df_tech = pd.DataFrame(tech_results, columns=["Technology", "Number of Jobs"])
df_location = pd.DataFrame(location_results, columns=["Location", "Number of Jobs"])

print("\nTechnology Jobs:")
print(df_tech)
print("\nLocation Jobs:")
print(df_location)
```
