### Objective: Scrape links, images, and table data from webpages using BeautifulSoup

```python
from bs4 import BeautifulSoup  
import requests                 

print("Libraries imported successfully!")
```

```python
# Download the contents of IBM's webpage
url = "http://www.ibm.com"
data = requests.get(url).text

# Create a soup object
soup = BeautifulSoup(data, "html.parser")

# Scrape all image URLs
print("Image URLs found on IBM.com:")
print("-" * 50)
for link in soup.find_all('img'):
    print(link.get('src'))
```

```python
# URL containing HTML table with color codes
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"

# Get the webpage content
data = requests.get(URL).text
soup = BeautifulSoup(data, "html.parser")

# Find the table
table = soup.find('table')

# Extract all rows
print("Color Name ---> Color Code")
print("-" * 40)
for row in table.find_all('tr'):
    cols = row.find_all('td')
    if len(cols) > 3:  # make sure row has enough columns
        color_name = cols[2].getText()
        color_code = cols[3].getText()
        print("{} ---> {}".format(color_name, color_code))
```
