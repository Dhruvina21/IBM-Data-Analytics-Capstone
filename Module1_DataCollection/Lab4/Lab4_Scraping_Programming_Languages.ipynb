# Lab 4: Scraping Programming Languages & Salaries
## IBM Data Analytics Capstone Project
### Objective: Scrape programming language names and average annual salaries, save to CSV


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

print("Libraries imported successfully!")
```

```python
# URL containing programming languages and salary data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"

# Download the webpage
data = requests.get(url).text

# Create soup object
soup = BeautifulSoup(data, "html.parser")

print("Webpage downloaded successfully!")
print("Page title:", soup.title.string)
```

```python
# Find the table
table = soup.find('table')

# Lists to store scraped data
languages = []
salaries = []

# Loop through all rows
for row in table.find_all('tr'):
    cols = row.find_all('td')
    if len(cols) > 0:
        language = cols[0].getText()
        salary = cols[3].getText()
        languages.append(language)
        salaries.append(salary)
        print(f"{language} ---> {salary}")
```

```python
# examine the first few rows to find correct column indices
for i, row in enumerate(table.find_all('tr')[:3]):
    cols = row.find_all('td')
    print(f"Row {i}:")
    for j, col in enumerate(cols):
        print(f"  Column {j}: {col.getText()}")
    print()
```

```python
# Lists to store scraped data
languages = []
salaries = []

# Loop through all rows skipping the header row
for row in table.find_all('tr')[1:]:  # skip row 0 (header)
    cols = row.find_all('td')
    if len(cols) > 0:
        language = cols[1].getText()
        salary = cols[3].getText()
        languages.append(language)
        salaries.append(salary)
        print(f"{language} ---> {salary}")
```

```python
# Create a DataFrame
df = pd.DataFrame({
    "Language": languages,
    "Average Annual Salary": salaries
})

print(df)

# Save to CSV
df.to_csv("popular-languages.csv", index=False)
print("\n popular-languages.csv saved successfully!")
```
