### Objective: Load, explore and summarize the Stack Overflow Developer Survey dataset


```python
import pandas as pd
print("Pandas version:", pd.__version__)
```

```python
# Load the dataset directly from the URL
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"

df = pd.read_csv(file_path)
print(" Dataset loaded successfully!")
```

```python
# Display the top 5 rows of the dataset
df.head()
```

```python
# Print the number of rows and columns
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
# Print the datatype of all columns
print(df.dtypes)
```

```python
# Print the mean age of the survey participants
print("Mean age of survey participants:")
print(df['Age'].value_counts())
print("\nNote: Age is stored as a range (object type), not a numeric value")
```

```python
# Print the number of unique countries
print("Number of unique countries:", df['Country'].nunique())
print("\nTop 10 countries by responses:")
print(df['Country'].value_counts().head(10))
```

## Key Findings from Dataset Exploration

- **Total Responses:** 65,457 developers worldwide
- **Total Columns:** 114 features
- **Countries Represented:** 185 unique countries
- **Top Country:** USA with 11,099 responses
- **Largest Age Group:** 25-34 years old (23,913 respondents)
- **Data Types:** Mix of int64, float64, and object columns
- **Dataset Source:** Stack Overflow Developer Survey (ODbL License)
