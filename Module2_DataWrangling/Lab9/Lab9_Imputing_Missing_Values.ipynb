# Lab 9: Imputing Missing Values
## IBM Data Analytics Capstone Project 
### Objective: Identify, impute missing values and normalize data in the Stack Overflow Developer Survey dataset

```python
import pandas as pd

# Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

print("Dataset loaded successfully!")
print("Shape:", df.shape)
df.head()
```

```python
# Task 1: Identify duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())

# Task 2: Remove duplicates
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
print("Duplicates removed!")
```

```python
# Complete missing values for ALL columns
missing_all = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_all_df = pd.DataFrame({
    'Missing Values': missing_all,
    'Percentage (%)': missing_percent.round(2)
}).sort_values('Missing Values', ascending=False)

# Show ALL columns with missing values
print(f"Total columns with missing values: {missing_all_df[missing_all_df['Missing Values'] > 0].shape[0]}")
print(f"Total columns with NO missing values: {missing_all_df[missing_all_df['Missing Values'] == 0].shape[0]}")
print("\nComplete missing values report:")
pd.set_option('display.max_rows', 120)
print(missing_all_df[missing_all_df['Missing Values'] > 0])
```

```python
# Task 5: Value counts for RemoteWork
print("RemoteWork value counts:")
print(df['RemoteWork'].value_counts())

# Task 6: Most frequent value
most_frequent_remotework = df['RemoteWork'].value_counts().idxmax()
print("\nMost frequent RemoteWork value:", most_frequent_remotework)
```

```python
# Task 7: Impute missing values in RemoteWork with majority value
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent_remotework)

print("Missing values after imputation:", df['RemoteWork'].isnull().sum())
print("RemoteWork imputed successfully!")
print("\nUpdated value counts:")
print(df['RemoteWork'].value_counts())
```

```python
# Task 8: Check compensation-related columns
comp_cols = ['CompTotal', 'ConvertedCompYearly']

for col in comp_cols:
    print(f"\n{col}:")
    print(f"  Missing values: {df[col].isnull().sum()}")
    print(f"  Description:\n{df[col].describe()}")
```

## Summary

- **Missing Values:** 109 out of 114 columns had missing data
- **RemoteWork:** 10,631 missing values imputed with majority value 
  → "Hybrid (some remote, some in-person)"

### Compensation Analysis:
