# Lab 7: Removing Duplicates & Handling Missing Values
## IBM Data Analytics Capstone Project 
### Objective: Remove duplicates, handle missing values, and normalize compensation data

```python
import pandas as pd

# Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

print("Dataset loaded successfully!")
print("Shape:", df.shape)
df.head()
```

```python
# Count duplicate rows
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)

# Display first few duplicates if any
if duplicate_count > 0:
    print("\nFirst few duplicate rows:")
    print(df[df.duplicated()].head())
else:
    print(" No duplicate rows found in this dataset!")
```

```python
# Remove duplicates (verification step)
df_clean = df.drop_duplicates()

print("Original shape:", df.shape)
print("After drop_duplicates shape:", df_clean.shape)
print("Rows removed:", df.shape[0] - df_clean.shape[0])
print(" Dataset is already free of duplicates!")
```

```python
# Identify missing values for all columns
missing_values = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percent
})

# Show only columns with missing values
missing_df = missing_df[missing_df['Missing Values'] > 0].sort_values('Missing Values', ascending=False)

print("Columns with missing values:")
print(missing_df.head(20))
```

```python
# Check missing values in EdLevel specifically
print("Missing values in EdLevel:", df['EdLevel'].isnull().sum())
print("\nEdLevel value counts:")
print(df['EdLevel'].value_counts())

# Impute with most frequent value
most_frequent_edlevel = df['EdLevel'].value_counts().idxmax()
print("\nMost frequent EdLevel:", most_frequent_edlevel)

df['EdLevel'].fillna(most_frequent_edlevel, inplace=True)
print("Missing values in EdLevel after imputation:", df['EdLevel'].isnull().sum())
print(" EdLevel missing values imputed successfully!")
```

```python
# Check missing values in ConvertedCompYearly
print("Missing values in ConvertedCompYearly:", df['ConvertedCompYearly'].isnull().sum())
print("Total rows:", len(df))

# Handle missing values - fill with median (better than mean for salary data)
median_comp = df['ConvertedCompYearly'].median()
print("\nMedian CompYearly:", median_comp)

df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_comp)
print("Missing values after imputation:", df['ConvertedCompYearly'].isnull().sum())

# Summary statistics
print("\nConvertedCompYearly Summary:")
print(df['ConvertedCompYearly'].describe())
print(" Compensation data normalized successfully!")
```
